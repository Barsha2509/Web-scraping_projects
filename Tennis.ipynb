# Tennis rankings scraper

In this project, data about the tennis players (men and women) rankings was collected from the ESPN website.Requests and beautiful soup libraries were used to collect the data from this website. All the collected data was stored in Python's lists. These lists were then converted into a dataframe which was then saved as a csv file. Finally, Tableau was used for data visualization. The link to the Tableau story is here
https://public.tableau.com/app/profile/bt4661/viz/Tennis2021/Story1

```python
import requests
import pandas as pd
import numpy as np
```

```python
url='https://www.espn.com/tennis/rankings'
r=requests.get(url).text
df=pd.read_html(r)
```

```python
from bs4 import BeautifulSoup 
soup = BeautifulSoup(r.text, 'html5lib')
print(soup.prettify())
s=soup.json()
```

```python
cou=soup.find_all('img', class_='Image Logo Logo__sm')
```

```python
country=[]
for i in cou:
    count=i.get('title')
    country.append(count)
country
```

```python
data['Country']=country
```

```python
data.drop(['Unnamed: 1'],  axis=1,inplace=True)
data.head()
```

```python
URL=['https://www.espn.com/tennis/rankings','https://www.espn.com/tennis/rankings/_/type/wta']
for url in range(0,2):
    r=requests.get(URL[url]).text
    if URL[url]=='https://www.espn.com/tennis/rankings':
        df=pd.read_html(r)
        men=df[0]    
        soup=BeautifulSoup(r, 'html5lib')
        cou=soup.find_all('img', class_='Image Logo Logo__sm')
        country=[]
        for i in cou:
            count=i.get('title')
            country.append(count)
        men['Country']= country
            
        
    elif URL[url]=='https://www.espn.com/tennis/rankings/_/type/wta':
        df=pd.read_html(r)
        women=df[0]    
        soup=BeautifulSoup(r, 'html5lib')
        cou=soup.find_all('img', class_='Image Logo Logo__sm')
        country=[]
        for i in cou:
            count=i.get('title')
            country.append(count)
        women['Country']= country
            
    
    
```
