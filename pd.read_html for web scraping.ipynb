```python
import pandas as pd
import requests
```

```python
url='http://www.worldometers.info/gdp/gdp-by-country/' #url for the given webpage
```

```python
data=requests.get(url).text #call the api using the get method and store the
                        # output of the API call in a variable called data
```

```python
df_list=pd.read_html(data) #Read HTML tables into a list of DataFrame objects.
df_list
```

```python
df=df_list[0] #creating dataframe from the list of dataframe objects
df.head()
```
